```python
   import requests

   # Definiowanie URL do API
   url = "http://danepubliczne.imgw.pl/api/data/hydro/"

   try:
       # Wykonanie żądania GET
       response = requests.get(url)

       # Sprawdzenie, czy żądanie zakończyło się sukcesem (kod statusu 200)
       if response.status_code == 200:
           # Pobranie danych w formacie JSON
           data = response.json()

           # Wyświetlenie pobranych danych
           print(data)
       else:
           print(f"Błąd: {response.status_code}")

   except requests.exceptions.RequestException as e:
       # Obsługa błędów
       print(f"Wystąpił błąd: {e}")
   
```

```python
   import requests
   import pandas as pd

   # Definiowanie URL do API
   url = "http://danepubliczne.imgw.pl/api/data/hydro/"

   try:
       # Wykonanie żądania GET
       response = requests.get(url)

       # Sprawdzenie, czy żądanie zakończyło się sukcesem (kod statusu 200)
       if response.status_code == 200:
           # Pobranie danych w formacie JSON
           data = response.json()

           # Przekształcenie danych JSON na ramkę danych pandas
           df = pd.DataFrame(data)

           # Wyświetlenie pierwszych kilku wierszy ramki danych
           print(df.head())
       else:
           print(f"Błąd: {response.status_code}")

   except requests.exceptions.RequestException as e:
       # Obsługa błędów
       print(f"Wystąpił błąd: {e}")
   
```
