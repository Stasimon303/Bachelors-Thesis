```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Ścieżka do pliku shapefile
shapefile_path = "C:/Users/Stanislav/OneDrive/Pulpit/Data Science and Licencjat/Licencjat/Dane/05282016_Brazos_River_near_Hempstead_TX/05282016_Brazos_River_near_Hempstead_TX.shp"

# Wczytaj dane z pliku shapefile
gdf = gpd.read_file(shapefile_path)

# Wyświetl pierwsze kilka wierszy danych
print(gdf.head())

# Wyświetl podstawowe informacje o danych
print(gdf.info())

# Wyświetl dane na mapie
gdf.plot()
plt.title('Catchment Boundary for Station ID')
plt.show()

# Wyświetl współrzędne geometryczne
print(gdf['geometry'])

# Wyświetl atrybuty pliku
print(gdf.columns)

# Opcjonalnie: Przeanalizuj atrybuty
print(gdf.describe())
```
