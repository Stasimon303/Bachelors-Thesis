```python
import geopandas as gpd
import matplotlib.pyplot as plt

# cieka do pliku shapefile
shapefile_path = "C:/Users/Stanislav/OneDrive/Pulpit/Data Science and Licencjat/Licencjat/Dane/05282016_Brazos_River_near_Hempstead_TX/05282016_Brazos_River_near_Hempstead_TX.shp"

# Wczytaj dane z pliku shapefile
gdf = gpd.read_file(shapefile_path)

# Wywietl pierwsze kilka wierszy danych
print(gdf.head())

# Wywietl podstawowe informacje o danych
print(gdf.info())

# Wywietl dane na mapie
gdf.plot()
plt.title('Catchment Boundary for Station ID')
plt.show()

# Wywietl wsprzdne geometryczne
print(gdf['geometry'])

# Wywietl atrybuty pliku
print(gdf.columns)

# Opcjonalnie: Przeanalizuj atrybuty
print(gdf.describe())
```
